```python
from sklearn.datasets import fetch_california_housing #sklearn.datasets module in the scikitlearn library that provide utilities to load and generate a variety of datasets for ML practice
from sklearn.model_selection import train_test_split  #train_test_spit is the function from sklearn.model_selection used to randomly partion a dataset into training and testing subsets
import pandas as pd
```

```python

data=fetch_california_housing()
```

Convert to data frame :
the process of restructuring the data from various formtas(python lists, dictionaries,external files like csvs) into pandas data frame which two dimentional,tabular data structure used in python for data analysis

The code df = pd.DataFrame(data.data, columns=data.feature_names) creates a pandas DataFrame named df from existing data

*   pd.DataFrame(): create a two-dimensional, tabular data structure with labeled rows and columns, similar to a spreadsheet or SQL table.
*   data.data:his provides the actual raw data (likely a NumPy array) that will populate the rows and columns of the new DataFrame.


*   columns=data.feature_names:: This argument explicitly assigns labels (names) to each column of the DataFrame using the values from data.feature_names. This makes the data human-readable and easier to work with, as you can refer to columns by name rather than just their numerical index







```python
#df=pd.DataFrame(data.data,columns=data.feature_names)
df = pd.DataFrame(data.data, columns=data.feature_names)
df['price']=data.target
```
